```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('C:\\Users\\Priyanka\\Desktop\\SL Project\\vaccination-coverage-analysis-main\\Cleaned-data-vc.csv', index_col = 0)
```

```python
df.head()
```

```python
sns.set(font_scale=1.4)
df['sampleSize'].plot(kind='hist', figsize=(10, 10));
plt.xlabel("Sample Size", labelpad=14)
plt.ylabel("No of rows with that sampleSize value", labelpad=14)
plt.title("Range of sampleSize", y=1.015, fontsize=22);
```

```python
sns.set(font_scale=1.4)
df['confidenceInterval'].plot(kind='hist', figsize=(10, 10));
plt.xlabel("Confidence Interval", labelpad=14)
plt.ylabel("No of rows with that confidenceInterval value", labelpad=14)
plt.title("Range of confidenceInterval", y=1.015, fontsize=22);
```

```python
fig, ax = plt.subplots(figsize=(10, 6))
ax.scatter(x = df['confidenceInterval'], y = df['value'])
plt.xlabel("Confidence Value")
plt.ylabel("Value")
plt.show()
```
